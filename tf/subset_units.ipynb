```python
import os
import sys 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set(style = 'white', font_scale=1.3)
%autosave 5
np.random.seed(0)

# constants
file_ext = '.npz'
n_conv_layers = 5 
n_fc_layers = 3 
n_layers = n_conv_layers + n_fc_layers
layer_names = ['conv%d' % (l+1) for l in range(n_conv_layers)] \
+ ['fc%d' % (l+1) for l in range(n_fc_layers)]

# params 
condition = 'train'
# subj_id = 0
n_units_select = 5000
n_subjs = 5
```

```python
def take_random_subset(layer_act, n_units_select): 
    """take a random subset of all units 
    """
    n_units_total = np.shape(layer_act)[-1]
    if n_units_select > n_units_total: 
        return layer_act
    else:
        units_select_idx = np.random.choice(
            n_units_total, size = n_units_select, replace=False)
        units_select_idx = sorted(units_select_idx)
        layer_act_subset = layer_act[:,:,units_select_idx]
    return layer_act_subset
```

```python
out_root = os.path.join('/tigress/qlu/data/fmri_imagenet/data/alex/max_u%d/layers' % n_units_select)
subj_id = 0 
for subj_id in range(n_subjs):
    
    print('subject id = %d' % subj_id)
    # activation file path 
    ckpt_path = '/tigress/qlu/logs/ILSVRC2012/ckpt-alexnet%.2d' % subj_id
    # main activation files path
    act_path = os.path.join(ckpt_path, 'acts', condition)

    # l = 6
    for l in range(n_layers): 
        # input path
        layer_act_path = os.path.join(act_path, layer_names[l] + file_ext)
        # output path
        out_path = os.path.join(out_root, 'cnn%d/s%d_%s' % (l+1, subj_id, condition)) 
        if not os.path.exists(os.path.dirname(out_path)): 
            os.makedirs(os.path.dirname(out_path))
            print('mkdir: ',os.path.dirname(out_path))
        # load the data 
        layer_act_file = np.load(layer_act_path)
        layer_act = layer_act_file['acts']
        # fetch random activities 
        layer_act_subset = take_random_subset(layer_act, n_units_select)
        layer_act_subset_ur = np.reshape(
            layer_act_subset, (-1, np.shape(layer_act_subset)[-1]))
        # save the data 
        np.savez(out_path, X=layer_act_subset_ur)
        print('\tREAD: %s' % layer_act_path)
        print('\tSAVE: %s' % out_path)
```
